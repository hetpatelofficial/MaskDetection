```python
TrainImgPath='D:/Python/Projects/Mask Detection/1/DATA/train'

trainidg=ImageDataGenerator(shear_range=0.1, zoom_range=0.1,horizontal_flip=True)
testidg = ImageDataGenerator()

#Generating the training data.

trainingset=trainidg.flow_from_directory(TrainImgPath, target_size=(64,64),batch_size=32,class_mode='categorical') 
```

```python
testset=testidg.flow_from_directory(TrainImgPath,target_size=(64,64),batch_size=32, class_mode='categorical')
testset.class_indices
```

```python
trainclasses=trainingset.class_indices
```

```python
#Storing faces and numeric tag for references
ResultMap={}

for faceValue,faceName in zip(trainclasses.values(),trainclasses.keys()):
  ResultMap[faceValue]=faceName
```

```python
import pickle
with open("ResultMap.pkl",'wb') as fileWriteStream:
    pickle.dump(ResultMap, fileWriteStream)
print("Mapping of Faces and IDs", ResultMap)
OutputNeurons=len(ResultMap)
print("Number of Neurous", OutputNeurons)
```

```python
from keras.models import Sequential
from keras.layers import Convolution2D
from keras.layers import MaxPool2D
from keras.layers import Flatten
from keras.layers import Dense
import time
import pickle
import numpy as np
import cv2
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
```

```python
demo_model1=Sequential()

demo_model1.add(Convolution2D(32,kernel_size=(5,5), strides=(1,1), input_shape=(64,64,3),activation='relu'))
demo_model1.add(MaxPool2D(pool_size=(2,2)))

demo_model1.add(Convolution2D(32,kernel_size=(5,5), strides=(1,1),activation='relu'))
demo_model1.add(MaxPool2D(pool_size=(2,2)))

demo_model1.add(Convolution2D(64,kernel_size=(5,5), strides=(1,1),activation='relu'))
demo_model1.add(MaxPool2D(pool_size=(2,2)))

demo_model1.add(Flatten())

demo_model1.add(Dense(200, activation='relu'))

demo_model1.add(Dense(OutputNeurons,activation='softmax'))


demo_model1.compile(loss='categorical_crossentropy', optimizer='adam')
```

```python
demo_model1.fit(trainingset, steps_per_epoch=20, epochs=20, validation_data=testset, validation_steps=10)

```
